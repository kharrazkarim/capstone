```python
# Clustering with spark 
#import pandas library, KMEANS and numpy to manipulate metrics
#import data using pandas
import pandas as pd
from pyspark.mllib.clustering import KMeans, KMeansModel
from numpy import array
```

```python
#import ad-clicks data using pandas and then cleaning it by removing white space
adclicks_df = pd.read_csv('file:///home/cloudera/Downloads/capston-project/flamingo-data/ad-clicks.csv')
adclicks_df = adclicks_df.rename(columns=lambda x: x.strip())
adclicks_df.head(10)
```

```python
#import gameclicks data using pandas, cleaning it by removing white space and display the first 10
gameclicks_df = pd.read_csv('file:///home/cloudera/Downloads/capston-project/flamingo-data/game-clicks.csv')
gameclicks_df = gameclicks_df.rename(columns=lambda x: x.strip())
gameclicks_df.head(10)
```

```python
#import buyclicks data using pandas, cleaning it by removing white space and display the first 10 item 
buyclicks_df = pd.read_csv('file:///home/cloudera/Downloads/capston-project/flamingo-data/buy-clicks.csv')
buyclicks_df = buyclicks_df.rename(columns=lambda x: x.strip())
buyclicks_df.head(10)
```

```python
#Ici on ajoute une colonne qui contient des 1 a gameclick data afin de savoir chaque user le nombre total des clicks par user 
gameclicks_df['clickCount'] = 1
gameclicks_df.head(10)
```

```python
#Ici on ajoute une colonne qui contient des 1 a adclicks_df afin de savoir chaque user le nombre total des clicks sur les pubs par user 
adclicks_df['adCount'] = 1
adclicks_df.head(10)
```

```python
#Dans notre analyse en s`interesse au prix afin de savoir le pouvoir d`achat de chaque user
user_achat = buyclicks_df[['userId','price']]
user_achat.head(10)
```

```python
#Ici on s`intersse qu`au nombre des clicks pour les 2 tables adclick et gameclick
user_adclicks = adclicks_df[['userId','adCount']]
user_gameclicks = gameclicks_df[['userId','clickCount']]
user_adclicks.head(10)
```

```python
user_gameclicks.head(10)
```

```python
#ici on calcul le total des clicks sur les pubs par user
user_total_ads = user_adclicks.groupby('userId').sum()
user_total_ads.head(10)
```

```python
user_total_ads = user_total_ads.reset_index()
user_total_ads.columns = ['userId', 'totalAdClicks']
user_total_ads.head(10)
```

```python
#on fait la meme chose pour les gameclicks
user_total_game_clicks = user_gameclicks.groupby('userId').sum()
user_total_game_clicks = user_total_game_clicks.reset_index()
user_total_game_clicks.columns = ['userId', 'totalGameClicks']
user_total_game_clicks.head(10)
```

```python
#afin de savoir le revenue total pour chaque user on fait la somme sur la colonne price de user_achat
user_total_revenue = user_achat.groupby('userId').sum()
user_total_revenue = user_total_revenue.reset_index()
user_total_revenue.columns = ['userId', 'totalAchat'] 
user_total_revenue.head(10)
```

```python
#ici on fait la jointure sur l attribut userId des 3 tables qu`on a et on va trouver pour chaque user le nombre de click sur les pubs sur le jeu et le total d`achat
join_df = user_total_ads.merge(user_total_game_clicks, on='userId')
join_df = join_df.merge(user_total_revenue, on='userId')
join_df.head(10)
```
